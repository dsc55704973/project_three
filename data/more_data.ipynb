```python
Cos = pd.read_csv('cosmetics.csv')
nfl = pd.read_csv('predictor_data.csv')
Cos.head()
```

```python
nfl = nfl.rename(columns = {'Date': 'date'})
nfl.head()
```

```python
Cos = Cos.rename(columns = {"Value": "Cosmetics_index", "Date": "date"})
Cos.head()
```

```python
import datetime
nfl['date'] = pd.to_datetime(nfl['date'], format='%m/%d/%Y')
nfl.head()
```

```python
Cos['date'] = pd.to_datetime(Cos['date'], format='%Y-%m-%d')
Cos.head()
```

```python
merged = nfl.merge(Cos, left_on='date', right_on='date')
len(merged)
merged.head()
```

```python
print(Cos.dtypes, nfl.dtypes)
```

```python
merged.to_csv(r'merged.csv', index = False)
```
