```python
import pandas as pd
```

```python
combined_data = pd.read_csv('Combined_data.csv')
merged = pd.read_csv('merged.csv')
```

```python
combined_data.head()
```

```python
merged.dtypes
merged.head()
```

```python
combined_data.head()
```

```python
combined_data['Date'] = pd.to_datetime(combined_data['Date'], format='%m/%d/%y')
combined_data.head()
```

```python
merged['date'] = pd.to_datetime(merged['date'], format='%Y-%m-%d')
merged.head()
```

```python
all_data = combined_data.join(merged.set_index('date'), on='Date')
all_data.head()
```

```python
all_data.to_csv(r'all_data.csv', index = False)
```
